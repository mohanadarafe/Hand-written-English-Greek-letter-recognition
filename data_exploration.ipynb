```python
import numpy as np
import sklearn
import matplotlib.pyplot as plt
from collections import defaultdict


english_training_filename = './data/Assig1-Dataset/train_1.csv'
english_test_filename = './data/Assig1-Dataset/test_with_label_1.csv'
english_validate_filename = './data/Assig1-Dataset/val_1.csv'

greek_training_filename = './data/Assig1-Dataset/train_2.csv'
greek_test_filename = './data/Assig1-Dataset/test_with_label_2.csv'
greek_validate_filename = './data/Assig1-Dataset/val_2.csv'
```

```python
eng_data = np.loadtxt(english_training_filename, dtype=np.int,delimiter=',',skiprows=1)
greek_data = np.loadtxt(greek_training_filename, dtype=np.int,delimiter=',',skiprows=1)
```

```python
eng_letters = sorted('qwertyuiopasdfghjklzxcvbnm'.upper())
greek_letters = ['π', 'α', 'β', 'σ', 'γ', 'δ', 'λ', 'ω', 'µ', 'ξ']

def plot_letter_count(data,letters):
    letter_count = defaultdict(int)

    for image in data:
        letter_count[letters[image[-1]]] += 1

    letter_count = dict(letter_count)
    letters = sorted(letter_count.keys())
    sorted_letter_count = {}

    for letter in letters:
        sorted_letter_count[letter] = letter_count[letter]

    letter_count = sorted_letter_count
    print(letter_count)

    plt.bar(list(letter_count.keys()),list(letter_count.values()),width=0.9,color='g')
    plt.xlabel('letters')
    plt.ylabel('occurences')
    plt.show()
    return letter_count
```

```python
eng_letter_count = plot_letter_count(eng_data, eng_letters)
eng_letter_count.keys(),eng_letter_count.values()
```

```python
greek_letter_count = plot_letter_count(greek_data,greek_letters)
greek_letter_count.keys(),greek_letter_count.values()
```

```python
def showimg(data):
    fig = plt.figure(figsize=(32/4,32/4))
    for i in range(1,31):
        img = data[i,:-1].reshape(32,32)
        fig.add_subplot(5,6,i)
        plt.imshow(img)
```
